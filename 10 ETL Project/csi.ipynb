```python
import pandas as pd
```

```python
filePath = 'Resources/csi.json'
```

```python
df = pd.read_json(filePath, encoding='UTF-8')
```

```python
df.head()
```

```python
df.columns
```

```python
newdf = df[['geopoliticalarea', 'iso_code', 'destination_description', 'travel_embassyAndConsulate', 'entry_exit_requirements', 'local_laws_and_special_circumstances', 'safety_and_security', 'health']]
newdf.head()
```

```python
newdf = newdf.rename(columns={'iso_code': 'country_code', 
                              'geopoliticalarea': 'country_name', 
                              'destination_description': 'country_desc', 
                              'travel_embassyAndConsulate': 'embassy_consulate_usa', 
                              'entry_exit_requirements': 'entry_exit_requirements', 
                              'local_laws_and_special_circumstances': 'laws_circumstances', 
                              'safety_and_security': 'safety_and_security', 
                              'health': 'health_circumstances'})

newdf.head()
```

```python
newdf.columns
```

```python
# https://stackoverflow.com/questions/33599156/converting-html-in-a-pandas-dataframe-column-which-was-read-from-a-csv-file-to-p
# https://jakevdp.github.io/PythonDataScienceHandbook/04.13-geographic-data-with-basemap.html
```

```python
from bs4 import BeautifulSoup as bs
```

```python
bs(df['travel_embassyAndConsulate'][0], 'html.parser').text.replace('\xa0', '')
```

```python
cnt = 0
for x in newdf['country_desc']:
    newdf['country_name'][cnt] = bs(newdf['country_name'][cnt], 'html.parser').text.replace('\xa0', '')
    newdf['country_desc'][cnt] = bs(newdf['country_desc'][cnt], 'html.parser').text.replace('\xa0', '')
    newdf['embassy_consulate_usa'][cnt] = bs(newdf['embassy_consulate_usa'][cnt], 'html.parser').text.replace('\xa0', '')
    newdf['entry_exit_requirements'][cnt] = bs(newdf['entry_exit_requirements'][cnt], 'html.parser').text.replace('\xa0', '')
    newdf['laws_circumstances'][cnt] = bs(newdf['laws_circumstances'][cnt], 'html.parser').text.replace('\xa0', '')
    newdf['safety_and_security'][cnt] = bs(newdf['safety_and_security'][cnt], 'html.parser').text.replace('\xa0', '')
    newdf['health_circumstances'][cnt] = bs(newdf['health_circumstances'][cnt], 'html.parser').text.replace('\xa0', '')
    cnt = cnt + 1

newdf.head()
```

```python
import config as config
from sqlalchemy import create_engine
# creating connection url and connectiong to Postgres DB
connURL = f"postgresql://{config.mysql['user']}:{config.mysql['password']}{config.mysql['url']}"
engine = create_engine(connURL)
```

```python
engine.table_names()
```

```python
newdf.to_sql(name='country_travel', con=engine, if_exists='append', index=False)
```

```python
import pandas as pd
```

```python
filePath = 'Resources/European Backpackers Index 2018.xlsx'
```

```python
df = pd.read_excel(filePath)
```

```python
df.head()
```

```python
df.columns
```

```python
newdf = df[['Country', 'City', 'Best cheap hostel',
       'Local Currency Code', 'Rank', 'Total USD', 'Attractions (USD)',
       'Cost Per Night (Hostel) (USD)', 'Drinks Entertainment (USD)',
       'Meals (USD)', 'Transportation (USD)']]
newdf.head()
```

```python
newdf = newdf.rename(columns={
        'Country': 'country_name',
        'City': 'city_name',
        'Best cheap hostel': 'hostel_name',
        'Local Currency Code': 'local_currency',
        'Rank': 'rank',
        'Total USD': 'total_cost',
        'Attractions (USD)': 'attractions_cost',
        'Cost Per Night (Hostel) (USD)': 'per_night_hostel',
        'Drinks Entertainment (USD)': 'drinks_entertainment',
        'Meals (USD)': 'meals',
        'Transportation (USD)': 'transportation'
})

newdf.head(60)
```

```python
import config as config
from sqlalchemy import create_engine
# creating connection url and connectiong to Postgres DB
connURL = f"postgresql://{config.mysql['user']}:{config.mysql['password']}{config.mysql['url']}"
engine = create_engine(connURL)
```

```python
engine.table_names()
```

```python
newdf = newdf.drop(newdf.index[6])
newdf = newdf.drop(newdf.index[32])
newdf = newdf.drop(newdf.index[46])
newdf = newdf.drop(newdf.index[47])
newdf = newdf.drop(newdf.index[47])
newdf.head()
```

```python
newdf.to_sql(name='hostel_travel', con=engine, if_exists='append', index=False)
```
